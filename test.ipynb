```python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv("D:\LLM_Practice\llm.env"))
openai_api_key = os.environ["OPENAI_API_KEY"]
```

```python
import streamlit as st
import sqlite3
from langchain_openai import OpenAI
from langchain.chains import RetrievalQA
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
from PyPDF2 import PdfReader
```

```python
# Read the files
reader = PdfReader("110BENG-RAG+APP_+PDF.pdf")
formatted_documents = []
for page in reader.pages:
    formatted_documents.append(page.extract_text())
```

```python
#Split the text
text_splitter = CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap = 0
)

docs = text_splitter.create_documents(formatted_documents)
```

```python
#create embeddings
embeddings = OpenAIEmbeddings()
```

```python
#Store the embeddings

store = FAISS.from_documents(docs,embeddings)
```

```python
#Create retreival Chain
retrieval_chain = RetrievalQA.from_chain_type(
        llm=OpenAI(temperature=0),
        chain_type="stuff",
        retriever=store.as_retriever()
    )
```

```python
Query = "Summarise the document"
```

```python
retrieval_chain.run(Query)
```
